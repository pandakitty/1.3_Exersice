```python
# Import libraries
import numpy as np
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
```

```python
# create model
y = np.array ([-3, -1, -2, 1, -1, 2, 1, 3, 4, 3, 4, 5])
yhat = np.array([-2, 1, 1, 0, -1, 1, 2, 2, 3, 3, 3, 5])
x = list(range(len(y)))
```

```python
# Visualize teh dataset
# Check differences
plt.scatter(x, y, color="purple", label="original")
plt.plot(x, yhat, color="green", label="prediction")
plt.legend()
plt.show()
```

```python
# Metrics by Calculation
```

```python
d = y - yhat
mse_f = np.mean(d**2)
mae_f = np.mean(abs(d))
rmse_f = np.sqrt(mse_f)
r2_f = 1-(sum(d**2))/sum((y-np.mean(y))**2)

print("Results by manual calculation:")
print("Mean Absolute Error:", mae_f)
print("Mean Squared Error:", mse_f)
print("Root Mean Squared Error:", rmse_f)
print ("R-Squared:", r2_f)
```
